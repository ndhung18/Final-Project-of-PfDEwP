# Làm sạch và tiền xử lý dữ liệu

## Tổng quan dữ liệu thô

```python
import pandas as pd
import glob
import json
import re
```

### Ghép các file csv cào được thành file duy nhất để thao tác

```python
folder_path = 'Data/Raw_Data/'
csv_files = glob.glob(folder_path + '*.csv')
```

```python
dfs = []

for file in csv_files:
    df = pd.read_csv(file)
    dfs.append(df)
    
df_concat = pd.concat(dfs, axis=0, ignore_index=True)
```

### Thông tin về dữ liệu thô

```python
df_concat.info()
```

## Tổ chức lại dữ liệu

### Loại bỏ các dữ liệu bị thiếu trong cột 'reactions', sau đó loại bỏ các dữ liệu bị trùng theo 'post_id'

```python
df_reactions = df_concat[df_concat['reactions'].notnull()]
df_no_duplicates = df_reactions.drop_duplicates(subset='post_id', keep='first')
```

### Sắp xếp lại dữ liệu theo thứ tự thời gian

```python
df_no_duplicates = df_no_duplicates.copy()
df_no_duplicates['timestamp'] = pd.to_numeric(df_no_duplicates['timestamp'], errors='coerce')
df_timesort = df_no_duplicates.sort_values(by='timestamp', ascending=False)

df_timesort
```

### Loại bỏ các cột không cần

```python
df = df_timesort.drop(["post_text", "shared_text", "original_text", "timestamp", 
                       "image_lowquality", "images", "images_description", 
                       "images_lowquality", "images_lowquality_description", 
                       "video", "video_duration_seconds", "video_height", 
                       "video_id", "video_quality", "video_size_MB", "video_thumbnail", 
                       "video_watches", "video_width", "image", "post_url", "link",
                       "links", "user_id", "username","user_url", "is_live", "factcheck",
                       "shared_post_id", "shared_time", "shared_user_id", "shared_username", 
                       "shared_post_url", "available", "reactors","w3_fb_url", "with", "page_id",
                       "sharers", "image_id", "image_ids", "was_live", "fetched_time", "start_url", 
                       "video_ids", "videos", "header", "likes"], axis=1, inplace=False)
df.info()
```

## Xử lý dữ liệu trên từng cột

### Xử lý giá trị None trong cột 'text'

```python
df['text'] = df['text'].fillna('')
df.info()
```

### Chuyển kiểu dữ liệu của cột 'time' thành datetime

```python
df['time'] = pd.to_datetime(df['time'])
df.info()
```

### Xử lý cột 'reactions'

```python
df['reactions']
```

#### Tách 7 loại thông tin bao gồm ['like', 'love', 'care', 'haha', 'wow', 'sad', 'angry'] thành các cột riêng

```python
reaction_columns = ['like', 'love', 'care', 'haha', 'wow', 'sad', 'angry']
def seperate_reaction(reactions):
    reactions_dict = json.loads(reactions.replace("'", "\""))
    dict = {key: reactions_dict.get(key, 0) for key in reaction_columns}
    return pd.Series(dict)

df[reaction_columns] = df['reactions'].apply(seperate_reaction)
```

```python
df[reaction_columns]
```

#### Loại bỏ cột reactions do không còn dùng nữa

```python
df = df.drop(['reactions'], axis=1, inplace=False)
```

### Xử lí cột 'comments_full'

```python
def extract_comments(text):
    pattern = r"\{[^{}]*\}"
    comment_strings = re.findall(pattern, text)
    cmts = []
    for comment_str in comment_strings:
        try:
            cmt = eval(comment_str)
            cmts.append(cmt)
        except:
            continue
    return cmts
```

```python
def cmts_df(df):
    df_result = {"comment_id": [], "comment_text": [], "comment_time": [], "post_id": []}
    for _,row in df.iterrows():
        for comment in extract_comments(row["comments_full"]):
            try:
                df_result['comment_id'].append(comment['comment_id'])
                df_result['comment_text'].append(comment['comment_text'])
                df_result['comment_time'].append(comment['comment_time'])
                df_result['post_id'].append(row.post_id)
            except:
                continue
    
    return pd.DataFrame(df_result)
```

```python
df_cmts = cmts_df(df)
df_cmts
```

```python
df_cmts.to_csv('Data/comment.csv', index = False)
```

#### Xoá bỏ cột 'comments_full' do không còn dùng đến nữa

```python
df = df.drop(['comments_full'], axis=1, inplace = False)
```
